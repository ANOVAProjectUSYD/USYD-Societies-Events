```python
import pandas as pd
import matplotlib as mlt
import matplotlib.pyplot as plt
import seaborn as sns
import pygal
```

```python
data = pd.read_csv('final_societies_facebook.csv')
```

```python
# First sort the dataset in descending order by number of likes on their page.
data = data.sort_values(by = ['Likes'], ascending=False)
```

```python
data.head()
```

Now plotting to see the frequency of the types of societies at USYD.

```python
plt.figure(figsize=(20, 10)) # this creates a figure 8 inch wide, 4 inch high
sns.set(style="darkgrid")
ax = sns.countplot(x="Category", data=data).set_title("Types of Societies on Campus")
plt.tight_layout()
plt.show()
```

```python
# Total number of likes for each TYPE of society and display. Use Groupby function to do this.
obj = data.groupby('Category')
data2 = obj.sum()
plt.figure(figsize=(20, 10)) # this creates a figure 8 inch wide, 4 inch high
ax = sns.barplot(x='Category',y='Likes', data=data2.reset_index())
ax.set_xlabel('Category')
ax.set_ylabel('Number of Likes')
ax.set_title('Break down of likes by category')
```

```python
plt.figure(figsize=(15, 7)) # this creates a figure 8 inch wide, 4 inch high

ax = sns.barplot(x="Society", y="Likes", data=data.head(20))
ax.set_xlabel('Society Name')
ax.set_ylabel('Number of Likes')
ax.set_title('Top 20 biggest societies')
plt.xticks(rotation=90)
plt.show()
```

```python
# Display which types of societies in top 20
plt.figure(figsize=(10, 5)) # this creates a figure 8 inch wide, 4 inch high
ax = sns.countplot(x="Category", data=data.head(20)).set_title('Biggest 20 societies by categories.')
```

We can see then that societies that are course related nearly account for half of the biggest 20 societies at USYD.
