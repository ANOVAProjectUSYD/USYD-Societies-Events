```python
import pandas as pd
import matplotlib as mlt
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.plotly as py
from plotly.graph_objs import *
```

```python
data = pd.read_csv('final_societies_facebook.csv')
```

Checking to see if this works.

```python
data.head()
```

Now plotting to see the frequency of the types of societies at USYD.

```python
plt.figure(figsize=(20, 10)) # this creates a figure 8 inch wide, 4 inch high
sns.set(style="darkgrid")
ax = sns.countplot(x="Category", data=data).set_title("Types of Societies on Campus")
plt.tight_layout()
plt.show()
```

```python
# Total number of likes for each TYPE of society and display.
```

```python
ax = sns.barplot(x="Society", y="Likes", data=data.head(20))
ax.set_xlabel('Society')
ax.set_ylabel('Number of Likes')
ax.set_title('Top 20 biggest societies')
```

```python
# Display which types of societies in top 20
ax = sns.barplot(x="Category", y="Likes", data=data.head(20))
ax.set_xlabel('Category')
ax.set_ylabel('Number of Likes')
ax.set_title('Top 20 biggest societies')
```
